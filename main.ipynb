```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu t tp Excel
df = pd.read_excel('evaluation_results.xlsx')

# Thit lp kch thc  th
plt.figure(figsize=(10, 6))

# V  th
plt.plot(df.index, df['Loss'], marker='o', label='Loss')
plt.plot(df.index, df['Accuracy'], marker='o', label='Accuracy')

# Thit lp nhn trc x v y
plt.xlabel('Model')
plt.ylabel('Value')

# Thit lp nhn cho trc x
labels = ['SmallDataCnn', 'SmallDataLSTM', 'SmallDataCnn_LSTM', 'BigDataCnn', 'BigDataLSTM', 'BigDataCnn_LSTM']
plt.xticks(df.index, labels)

# Thit lp nhn cho mi im d liu
for i, j in zip(df.index, df.iterrows()):
    plt.text(i, j[1]['Loss'], f'({j[1]["Loss"]:.3f}, {j[1]["Accuracy"]:.3f})', ha='center', va='bottom')

# Thit lp ch thch
plt.legend()

# Hin th  th phng to
plt.show()
```
