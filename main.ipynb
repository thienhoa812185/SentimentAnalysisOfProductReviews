```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp Excel
df = pd.read_excel('evaluation_results.xlsx')

# Thiết lập kích thước đồ thị
plt.figure(figsize=(10, 6))

# Vẽ đồ thị
plt.plot(df.index, df['Loss'], marker='o', label='Loss')
plt.plot(df.index, df['Accuracy'], marker='o', label='Accuracy')

# Thiết lập nhãn trục x và y
plt.xlabel('Model')
plt.ylabel('Value')

# Thiết lập nhãn cho trục x
labels = ['SmallDataCnn', 'SmallDataLSTM', 'SmallDataCnn_LSTM', 'BigDataCnn', 'BigDataLSTM', 'BigDataCnn_LSTM']
plt.xticks(df.index, labels)

# Thiết lập nhãn cho mỗi điểm dữ liệu
for i, j in zip(df.index, df.iterrows()):
    plt.text(i, j[1]['Loss'], f'({j[1]["Loss"]:.3f}, {j[1]["Accuracy"]:.3f})', ha='center', va='bottom')

# Thiết lập chú thích
plt.legend()

# Hiển thị đồ thị phóng to
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp Excel
df = pd.read_excel('evaluation_results.xlsx')

# Thiết lập kích thước đồ thị
plt.figure(figsize=(10, 6))

# Vẽ đồ thị cho cột "Loss" với màu đỏ
plt.plot(df.index, df['Loss'], marker='o', color='orange')

# Thiết lập nhãn trục x và y
plt.xlabel('Model')
plt.ylabel('Loss')

# Thiết lập nhãn cho trục x
labels = ['SmallDataCnn', 'SmallDataLSTM', 'SmallDataCnn_LSTM', 'BigDataCnn', 'BigDataLSTM', 'BigDataCnn_LSTM']
plt.xticks(df.index, labels)

# Thiết lập tên cho đồ thị
plt.title('Loss Comparison')

# Thêm giá trị trên mỗi điểm
for i, j in zip(df.index, df['Loss']):
    plt.text(i, j, f'{j:.3f}', ha='center', va='bottom')

# Hiển thị đồ thị
plt.show()
```
